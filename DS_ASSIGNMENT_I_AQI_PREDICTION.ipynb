# Data Balancing

```python
# Sample data
sample_data = {'AQI Level': ['Good', 'Poor', 'Very Poor', 'Severe', 'Moderate', 'Satisfactory'],}

# Create a DataFrame from the data
df1 = pd.DataFrame(sample_data)

# Count the instances in each AQI level class
class_counts = df1['AQI Level'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6), dpi=80)
colors = [palette[level] for level in class_counts.index]
plt.pie(class_counts, labels=class_counts.index, colors=colors, autopct='%1.2f%%', startangle=140)

# Add a title
plt.title('AQI Level Class Distribution')

# Show the plot
plt.show()
```

```python
# List of numerical variables to plot histograms
numerical_variables = {'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene'}

# Set up subplots for histograms
plt.figure(figsize=(12, 8))

# Create histograms for each numerical variable
for i, variable in enumerate(numerical_variables, start=1):
    plt.subplot(3, 4, i)  # Create a subplot for each variable
    plt.hist(df[variable], bins=30, color='blue', alpha=0.7)
    plt.title(variable)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Statistical analysis of the data (variance, standard deviation, range)

```python
# Initialize a DataFrame to store the statistics
statistics_df = pd.DataFrame(columns=['Attribute', 'Variance', 'Standard Deviation', 'Range'])

# Calculate statistics for each attribute
for attribute in numerical_variables:
    # Calculate variance
    variance = np.var(df[attribute], ddof=1)  # ddof=1 for sample variance
    
    # Calculate standard deviation
    std_deviation = np.std(df[attribute], ddof=1)  # ddof=1 for sample standard deviation
    
    # Calculate range
    data_range = np.ptp(df[attribute])
    
    # Append statistics to the DataFrame
    statistics_df = statistics_df.append({'Attribute': attribute, 'Variance': variance, 'Standard Deviation': std_deviation, 'Range': data_range}, ignore_index=True)

# Display the statistics
print(statistics_df)
```
