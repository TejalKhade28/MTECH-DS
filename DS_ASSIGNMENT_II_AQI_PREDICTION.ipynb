# Filling missing values with mean of each numeric column

```python
# Define a list of column names to process
cols = ['PM2.5', 'PM10', 'NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
# Iterate through the columns and fill missing values with the mean
for column in cols:
    # Calculate the mean of the current column
    col_mean = df[column].mean()
    
    # Fill missing values in the column with the mean
    df[column].fillna(col_mean, inplace=True)

# Print the DataFrame with missing values filled
print(df)
```

```python
df.isnull().sum()
```

# Filling missing values with mode of categorical column

```python
category = df['AQI_Bucket'].mode().values[0]

# Print the mode
print(f"Mode of 'AQI_Bucket': {category}")
```

```python
# Fill missing categorical values with the mode
df['AQI_Bucket'].fillna(df['AQI_Bucket'].mode()[0], inplace=True)
```

We apply normalization to the selected columns by using the transform method on the DataFrame, 
which normalizes only the specified columns.

```python
import pandas as pd
from sklearn.preprocessing import Normalizer

# Specify the columns you want to normalize
columns_to_normalize = ['PM2.5', 'PM10', 'NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']

# Create a Normalizer instance (you can choose L1 or L2 normalization)
normalizer = Normalizer(norm='l2')  # Use 'l1' for L1 normalization

# Apply normalization to the selected columns
df[columns_to_normalize] = normalizer.transform(df[columns_to_normalize])

# Print the DataFrame with selected columns normalized
print(df)
```

# Ordinal Encoding

```python
# Create a sample DataFrame with an ordinal column 'Size'
data = {'Satisfaction': ['Good','Poor','Very Poor','Severe','Moderate','Satisfactory']}
df1 = pd.DataFrame(data)

# Define an ordinal mapping
size_mapping = {'Good': 0, 'Poor': 1, 'Very Poor': 2,'Severe': 3,'Moderate':4,'Satisfactory':5}

# Apply ordinal encoding
df1['Satisfaction'] = df1['Satisfaction'].map(size_mapping)

# Print the DataFrame with ordinal encoding
print(df1)
```

# Custom encoding for nominal data

```python
# Create a sample DataFrame with a nominal column 'City'
data1 = {'City_encoded': ['Ahmedabad', 'Aizawl', 'Amaravati', 'Amritsar','Bengaluru','Bhopal','Brajrajnagar','Chandigarh','Chennai','Coimbatore','Delhi','Ernakulam','Gurugram','Guwahati','Hyderabad','Jaipur','Jorapokhar','Kochi','Kolkata','Lucknow','Mumbai','Patna','Shillong','Talcher','Thiruvananthapuram','Visakhapatnam']}
df2 = pd.DataFrame(data1)

# Define an ordinal mapping
city_mapping = {'Ahmedabad':1, 'Aizawl':2, 'Amaravati':3, 'Amritsar':4,'Bengaluru':5,'Bhopal':6,'Brajrajnagar':7,'Chandigarh':8,'Chennai':9,'Coimbatore':10,'Delhi':11,'Ernakulam':12,'Gurugram':13,'Guwahati':14,'Hyderabad':15,'Jaipur':16,'Jorapokhar':17,'Kochi':18,'Kolkata':19,'Lucknow':20,'Mumbai':21,'Patna':22,'Shillong':23,'Talcher':24,'Thiruvananthapuram':25,'Visakhapatnam':26}

# Apply ordinal encoding
df2['City_encoded'] = df2['City_encoded'].map(city_mapping)

# Print the DataFrame
print(df2)
```

```python
result_df = pd.concat([df, df1, df2], axis=1)

# Print the resulting DataFrame
print(result_df)
```
