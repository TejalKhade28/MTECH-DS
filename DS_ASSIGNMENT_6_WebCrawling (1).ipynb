```python
from csv import writer
```

```python

# all_sections = soup.find_all('div', class_='col-xs-12 col-sm-8 col-lg-10 remove_space')
with open("dataScraped.csv",'w',newline='',encoding="utf") as f:
    thewriter=writer(f)
    header=['Name','Designation','Research Area','Mobile Number','Office Number','Office_Email','Personal_Email','Website URL']
    thewriter.writerow(header)
    for result in results:
        name=result.find('h4',class_="author-name").text.replace('\n','')
        print("Name: ",name)
        prof_info = result.find('p').text
        pos = prof_info.find('Research Area: ')
        if 'Research Area:' in prof_info:
            designation=prof_info[:pos].strip()

            print("Designation: ",designation)
            research_data = prof_info.split('Research Area: ')[1]
            research_areas = [area.strip() for area in research_data.split(',')]          
        else:
            research_areas="Research area data not found."
        print("Research Areas: ",research_areas)
        designation_ = result.find('p').text
        print("Designation:",designation_)
        mobile_number = result.find('i', class_='lnr lnr-smartphone')
        office_number = result.find_all('i', class_='lnr lnr-smartphone')

        if mobile_number and 'Mobile No.' in mobile_number.next_sibling:
            mobile_number = mobile_number.find_next('strong').text.strip()
        else:
            mobile_number = "Mobile No. not found"

        if office_number:
            found_office = False
            for item in office_number:
                if 'Office No.' in item.next_sibling:
                    office_number = item.find_next('strong').text.strip()
                    found_office = True
                    break
            if not found_office:
                office_number = "Office No. not found"
        else:
            office_number = "Office No. not found"

        print(f"Mobile No.: {mobile_number}")
        print(f"Office No.: {office_number}")



        text = result.get_text()

        words = text.split()
        emails = [word for word in words if ('[at]' in word) or ('@' in word)]
    #     print("emails", emails)
        office_email = "Office Email not found"
        personal_email = "Personal Email not found"

        if len(emails) > 0:
            if len(emails) == 1:
                office_email = emails[0]
            else:
                office_email = emails[0]
                personal_email = emails[1]

        print(f"Office Email: {office_email}")
        print(f"Personal Email: {personal_email}")



        website_url = result.find('a', class_='collapsed')
        if website_url:
            website_url = website_url['href']
        else:
            website_url="No website found!!"
        print(f"Website URL: {website_url}")
        professorData=[name,designation,research_areas,mobile_number,office_number,office_email,personal_email,website_url]
        thewriter.writerow(professorData)
        print("********************************************************************************************************")
```

```python
import requests

api_endpoint = 'https://www.svnit.ac.in/web/department/computer/faculty.php'
# response = requests.get(api_endpoint)


try:
    response = requests.get(api_endpoint)
    print(response)
    if response.status_code == 200:
        data = response.json
        for item in data['items']:
            name = item['name']
            office_email = item['office_email']
            personal_email = item['personal_email']
            office_number = item['office_number']
            designation = item['designation']
            research_area = item['research_area']
            webpage_link = item['webpage_link']

            # Print or use the extracted information as needed
            print("Name:", name)
            print("Office Email:", office_email)
            print("Personal Email:", personal_email)
            print("Office Number:", office_number)
            print("Designation:", designation)
            print("Research Area:", research_area)
            print("Webpage Link:", webpage_link)
            print("******************************************")
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")

except requests.RequestException as e:
    print(f"Request failed: {e}")
```

```python
import requests

# Example API endpoint (replace this with the actual API endpoint)
api_endpoint = 'https://www.svnit.ac.in/web/department/computer/faculty.php'


# Make a GET request to the API endpoint
response = requests.get(api_endpoint)

if response.status_code == 200:
    data = response.json()  # Assuming the data is returned in JSON format

    # Process the data
    for item in data['items']:
        name = item['name']
        office_email = item['office_email']
        personal_email = item['personal_email']
        office_number = item['office_number']
        designation = item['designation']
        research_area = item['research_area']
        webpage_link = item['webpage_link']

        # Print or use the extracted information as needed
        print("Name:", name)
        print("Office Email:", office_email)
        print("Personal Email:", personal_email)
        print("Office Number:", office_number)
        print("Designation:", designation)
        print("Research Area:", research_area)
        print("Webpage Link:", webpage_link)
        print("******************************************")

else:
    print("Failed to fetch data from the API. Status code:", response.status_code)
```
